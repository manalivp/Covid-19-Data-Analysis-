```python
#importing necessory modules

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print('Modules are imported.')
```

```python
data = pd.read_csv('/home/Covid_case_time_series.csv')
data.head()
#importing covid19 dataset
```

```python
#Let's check the shape of the dataframe

data.shape
```

```python
#We will be collecting Daily Confirmed  Daily Recovered and Daily Deceased in variables as array.

X = data.iloc[61:,0] #Stores Date
Y = data.iloc[61:,1].values #Stores Daily Confirmed
R = data.iloc[61:,3].values #Stores Daily Recovered
D = data.iloc[61:,5].values #Stores Daily Deceased
```

```python
#Plotting Simple Plot
#In this case, we will pass the ‘X’ variable which has ‘Dates’ and ‘Y’ variable which has ‘Daily Confirmed’ to plot.

plt.plot(X,Y)

#We get a Simple Plot on the execution of above code which looks like this where X-axis has Dates and Y-axis has Number of Confirmed cases.
```

```python
#To have control over the aesthetics of the graph such as labels, titles, color and size we shall apply more functions as shown below.

plt.figure(figsize=(25,8))

#This creates a canvas for the graph where the first value ‘25’ is the width argument position and ‘8’ is the height argument position of the graph.

#Let’s create an object of the axes of the graph as ‘ax’ so it becomes easier to implement functions.
ax = plt.axes()
plt.grid(axis = 'x')
ax.grid(linewidth=0.4, color='white')

#The width of the grid lines can be adjusted by passing an argument ‘linewidth’ and changing its color by passing ‘color’ argument.

ax.set_facecolor("pink")

ax.set_xlabel('Date',size=25, color='#4bb4f2')

ax.set_ylabel('Number of Confirmed Cases\n', size=25,color='#4bb4f2')

#‘.set_facecolor’ lets you set the background color of the graph which over here is black.
#‘.set_xlabel’ and ‘.set_ylabel’ lets you set the label along both axes whose size and color can be altered.

ax.plot(X,Y,
        color='#1F77B4',
        marker='o',
        linewidth=4,
        markersize=15,
        markeredgecolor='#035E9B')

#Now we plot the graph again with ‘X’ as Dates and ‘Y’ as Daily Confirmed by calling plot function. The plotting line , marker and color can be altered by passing color, linewidth?—?to change color and adjust the width of plotting line and marker, markersize, markeredgecolor?—?to create marker which is the circle in this case, adjust the size of the marker and define marker’s edge color.
```

```python
#Still as we can see the dates are overlapping and the labels along axes are not clear enough. So now we are going to change the ticks of the axes and also annotate the plots.

plt.figure(figsize=(25,8))

ax = plt.axes()
#plt.grid(axis = 'x')
ax.grid(linewidth=0.5, color='#8f8f8f')

ax.set_facecolor("black")
ax.set_xlabel('Date',size=25, color='#4bb4f2')
ax.set_ylabel('Number of Confirmed Cases', size=25, color='#4bb4f2')

plt.xticks(rotation='vertical',size='20',color='black')
#For the Dates to not overlap with each other we are going to represent them vertically by passing ‘rotation = vertical’.

plt.yticks(size=20,color='black')
#To make the ticks easily readable we change the font color to white and size to 20.

plt.tick_params(size=20,color='white')
#‘.tick_params’ lets you alter the size and color of the dashes which look act like a bridge between the ticks and graph.

for i,j in zip(X,Y):
    ax.annotate(str(j),xy=(i,j+100),color='white',size='13')

#annotate the plotted points by running a for loop which plots at the plotted points. Any string passed will be plotted.
#The XY is the coordinates where the string should be plotted. +100 to j in XY coordinates so that the string doesn’t overlap with the marker.

ax.annotate('Second Lockdown 15th April',
            xy=(15.2, 860),
            xytext=(19.9,500),
            color='white',
            size='25',
            arrowprops=dict(color='white',
                            linewidth=0.025))
#To annotate an arrow pointing at a position in graph and its tail holding the string we can define ‘arrowprops’ argument along with its tail coordinates defined by ‘xytext’.

plt.title("COVID-19 IN : Daily Confirmed\n", size=50,color='#28a9ff')

ax.plot(X,Y, color='#1F77B4',
        marker='o',
        linewidth=4,
        markersize=15,
        markeredgecolor='#035E9B')

plt.show()
```

```python
#Pie Chart

#We’ll be plotting the Transmission Pie Chart to understand the how the virus is spreading based on Travel, Place Visit and Unknown reason.

#Initializing Dataset
slices = [62, 142, 195] #contains the values representing the size of each pie slice.
activities = ['Travel', 'Place Visit', 'Unknown']

#List slices based on which our Pie Chart will be divided and the corresponding activities are it’s valued.

#Plotting Pie Chart
cols=['#4C8BE2','#00e061','#fe073a']
exp = [0.2,0.02,0.02] #used to create an "explode" effect, where slices are offset from the center for emphasis
plt.pie(slices,
labels=activities,
textprops=dict(size=25,color='black'),
radius=2,
colors=cols,
autopct='%2.2f%%',
explode=exp,
shadow=True,
startangle=50)

plt.title('Virus spreading Transmission \n\n\n',color='#4fb4f2',size=40)
```

```python
#Bar Plot
#Now we are going to plot the most common type of graph which is a bar plot. Over here we will be plotting the district wise coronavirus cases for a state.

#Initializing Dataset
data = pd.read_csv('/home/District wise cases.csv')
data.head()
```

```python
re=data.iloc[:30,5].values #‘re’ stores Recovered corona patients count for all districts.
de=data.iloc[:30,4].values #‘de’ stores Deceased corona patients count for all districts.
co=data.iloc[:30,3].values #‘co’ stores Confirmed corona patients count for all districts.
x=list(data.iloc[:30,0]) #‘x’ stored District names.
```

```python
#Plotting Bar Chart
plt.figure(figsize=(25,10))
ax=plt.axes()

ax.set_facecolor('black')
ax.grid(linewidth=0.4, color='#8f8f8f')

plt.xticks(rotation='vertical', size='20', color='white') #ticks of X

plt.yticks(size='20',color='white')

ax.set_xlabel('\nDistrict',size=25, color='#4bb4f2')
ax.set_ylabel('No. of cases\n',size=25, color='#4bb4f2')

plt.tick_params(size=20,color='white')

ax.set_title('Maharashtra District wise breakdown\n', size=50,color='#28a9ff')

plt.bar(x,co,label='re')
plt.bar(x,re,label='re',color='green')
plt.bar(x,de,label='re',color='red')

for i,j in zip(x,co):
    ax.annotate(str(int(j)),
                xy=(i,j+3),
                color='white',
                size='15')

plt.legend(['Confirmed','Recovered','Deceased'], fontsize=20)
```
